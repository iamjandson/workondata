```python
#Author: Jandson Ara√∫jo
#Goal: find the best amount of clusters

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

%matplotlib inline

a = datasets.load_wine()

#a

df = pd.DataFrame(a['data'],columns= a['feature_names'])
#df
df.hist(figsize=(15,10));

```

```python
from sklearn.cluster import KMeans

X = df

intra_cluster = []
for k in range (2, 10):
  
  model = KMeans(n_clusters = k)
  model = model.fit(X)
  intra_cluster.append(model.inertia_)
intra_cluster

plt.figure(figsize=(10,5))
plt.plot(range(2,10), intra_cluster)
```

```python
X['labels'] = model.labels_
sns.pairplot(X, hue ='labels', vars=X.drop(columns='labels'))

#X['labels'] = model.labels_
#


sns.pairplot(X, hue='labels', vars=X.drop(columns='labels'))
```
